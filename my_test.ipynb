# Test the model

Loads the saved model and test it with some data.

```python
import tensorflow as tf

import os.path as path
DATADIR = path.join(path.dirname(path.dirname(".")), "data")

def datadir(*p: str) -> str:
    return path.join(DATADIR, *p)

model_path = datadir("model_rot_compiled")
print(model_path)

model = tf.keras.models.load_model(model_path)
```

```python
import pandas as pd

input_dataset = pd.read_pickle(datadir("input_rot.pkl"))
```

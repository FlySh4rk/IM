```python
from utils import rotate, draw_square, draw_single
import numpy as np
import matplotlib.pyplot as plt

orig = input_dataset.to_numpy()[0]
#orig = np.array([-1, -1, 1, -1, 1, 1, -1, 1])
x = rotate(orig, theta=np.pi / 3)
display(x)

fig = plt.figure()
fig.add_axes((0, 0, 1, 1))


def draw_squares(N=5):
    ax = fig.axes[0]
    for i in range(2 * N):
        x = rotate(orig, theta=i * np.pi / N)
        draw_square(ax, x, color="green")
    draw_square(ax, orig, color="red")


draw_single(model=model, fig=fig, row=x)
```
